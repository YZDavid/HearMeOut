## Create a bunch of posts
With a sample of some textual data

```python
INPUTS = []
```

```python
for file in os.listdir("summary_files"):
    with open("summary_files/" + file, 'r') as f:
        item = {}
        item[file] = f.read()
        INPUTS.append(item)
INPUTS
```

```python
for file in INPUTS:
    for input in file:
        print(input, ":", len(file[input]), "characters long")
```

```python
for i in range(len(INPUTS)):
    input = list(INPUTS[i].values())[0]
    post_response = requests.post(BASE + '/conversions', data=json.dumps({'input': input, 'audio': True}))
    print(post_response.status_code)
    try:
        print(post_response.json())
    except:
        pass
```

## Testing out accessing conversion by ID

```python
length = len(get_response.json())
length
```

```python
before_after = []
```

```python
for i in range(4, length):
    get_by_id = requests.get(BASE + '/conversions/' + str(i))
    print(get_by_id.status_code)
    try:
        response = get_by_id.json()
        d = {}
        d['id'] = response.get('id')
        d['before_length'] = len(response.get('input'))
        d['after_length'] = len(response.get('output'))
        before_after.append(d)
    except:
        pass
```

```python
get_by_id.json()
```

```python
for entry in before_after:
    percent_original = round((entry['after_length'] / entry['before_length']) * 100, 2)
    print(f"After conversion, it is {percent_original}% of the initial length")
```
